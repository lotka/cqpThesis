```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline

def calculateKmin(matrix):
	return 1/(np.trace((matrix.T*matrix)*(matrix.T*matrix))/np.power(np.trace(matrix.T*matrix),2))

def getSchmidt(array):
	array = np.sqrt(abs(array))
	M = np.matrix(array)
	Kmin = calculateKmin(M)
	return Kmin

def getPurity(array):
    return 1.0/getSchmidt(array)

def getSVDMode(array,mode):
    #array = array/array.sum()
    U, diagonal, V = np.linalg.svd(array, full_matrices=True,compute_uv=True)
    diagonalMatrix = np.matrix(np.zeros((U.shape[0],V.shape[0])))
    diagonalMatrix[mode,mode]=1
    UMatrix = np.matrix(U)
    VMatrix = np.matrix(V)
    res = UMatrix*diagonalMatrix*VMatrix
    return res

def SVDModeCleaner(array,mode):
    array = array/array.sum()
    U, diagonal, V = np.linalg.svd(array, full_matrices=True,compute_uv=True)
    diagonalMatrix = np.matrix(np.zeros((U.shape[0],V.shape[0])))
    for i in xrange(mode):
        diagonalMatrix[i,i]=diagonal[i]
    UMatrix = np.matrix(U)
    VMatrix = np.matrix(V)
    res = UMatrix*diagonalMatrix*VMatrix
    return res

def errorSNRMethod(jsi,averageNoise):
    #get snr
    tot = float(jsi.shape[0]*jsi.shape[1])*averageNoise
    snr = jsi.sum()/tot
    print 'SNR=' + str(snr)
    
    # calculate errors
    params = [-2.75924958e-06,2.47538803e-01,3.77048100e-01,2.15234476e-02,1.13323311e-07]
    def errorModel(x, a, b, c, d, e):
        return a + b*np.exp(-c*x) + d*np.exp(-e*x)
    
    error = errorModel(snr,*params)
    print 'Error is: ' + str(error)
    return error
```
