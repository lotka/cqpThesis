```python
# Initialisation
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
import scipy.ndimage
import scipy.optimize as optimization

seperableGauss = lambda x,y: np.exp(-(x)**2)*np.exp(-(y)**2)
seperableXY = lambda x,y: (x*y)/1000000000
seperableSin = lambda x,y: (np.sin(x)*np.sin(y))**2
seperableLorentz = lambda x,y: 1/((x**2 + 1)*(y**2 + 1))
nonSeperableGauss = lambda x,y: np.exp(-(x+y)**2)
nonSeperableXY = lambda x,y: (x+y)/10
nonSeperableSin = lambda x,y: (np.sin(x+y))**2
nonSeperableLorentz = lambda x,y: 1/((x+y)**2 + 1)
nonSeperableWeird = lambda x,y: nonSeperableGauss(x,y)+nonSeperableXY(x,y)+nonSeperableSin(x,y)+nonSeperableLorentz(x,y)

def calculateKmin(matrix):
	return 1/(np.trace((matrix.T*matrix)*(matrix.T*matrix))/np.power(np.trace(matrix.T*matrix),2))

def getSchmidt(array):
	array = np.sqrt(abs(array))
	M = np.matrix(array)
	Kmin = calculateKmin(M)
	return Kmin

def getPurity(array):
    return 1.0/getSchmidt(array)
```

<h2>Now lets try to recover the signal which has an SNR of roughly 1.5</h2>
We take 5 iterations with noise strength 0.0005, array size 200 and we need the 5th iteration of the noise add algorithm

```python
noiseStrength = 0.0005
arraySize = 200
iteration = 6
```

```python
def recoverFromNoise(function,noiseStrength,arraySize):
    xSize = arraySize
    ySize = xSize
    js0 = np.zeros((xSize,ySize))
    js1 = np.zeros((xSize,ySize))
    
    #for i in xrange(xSize):
    #    js0[i,i]=1
    
    
    dataLength = 20
    data0 = np.zeros((3,dataLength))
    
    for i in xrange(xSize):
        for j in xrange(ySize):
            x = float(i-xSize/2)/10.0
            y = float(j-ySize/2)/10.0
            js0[i,j]=function(x,y)
           
    
    js0 = js0/js0.sum()
            
    plt.figure()
    a = plt.imshow(js0)#,interpolation='none')
    a.axes.text(20,20, 'Purity = ' + str(round(getPurity(js0),3)),bbox={'facecolor':'white', 'pad':10})
    
    noise = float(iteration)/float(dataLength)*(np.random.rand(xSize,xSize))*noiseStrength
    js1 = js0 + noise
    
    plt.figure()
    a = plt.imshow(js1)#,interpolation='none')
    a.axes.text(20,20, 'Purity = ' + str(round(getPurity(js1),3)),bbox={'facecolor':'white', 'pad':10})
    
    SNR = js1.sum()/noise.sum()
    print 'SNR = ' + str(SNR)
    
    js2 = js1 - noise.max()
    for i in xrange(js2.shape[0]):
        for j in xrange(js2.shape[1]):
            if js2[i,j] < 0:
                js2[i,j] = 0
    
    js2 = scipy.ndimage.gaussian_filter(js2,3)
    
    plt.figure()
    a = plt.imshow(js2)#,interpolation='none')
    a.axes.text(20,20, 'Purity = ' + str(round(getPurity(js2),3)),bbox={'facecolor':'white', 'pad':10})
```
